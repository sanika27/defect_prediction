```python
#All commits
from pydriller import RepositoryMining
repo_updated = "/Users/sanika/Downloads/Master Thesis/repo_updated/pytest"

count = 0
for commit in RepositoryMining(repo_updated).traverse_commits():
    count += 1
    print(count, commit.hash)
            
```

```python
#Bug inducing commits given bug fixed commits
from pydriller.git_repository import GitRepository

repo_updated = "/Users/sanika/Downloads/Master Thesis/repo_updated/pytest"
gr5 = GitRepository("/Users/sanika/Downloads/Master Thesis/repo_updated/pytest")

count5 = 0

for commit5 in RepositoryMining(repo_updated,
                                   only_modifications_with_file_types=['.py']).traverse_commits():
    if "fixed" in commit5.msg:
        commit6 = gr5.get_commit(commit5.hash)
        buggy_commits5 = gr5.get_commits_last_modified_lines(commit6) #get bug inducing commits
        count5 += 1
        print(count5, buggy_commits5)
```
